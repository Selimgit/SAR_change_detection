```python
from SAR_change_detector import detect_changes
import matplotlib.pyplot as plt
import tifffile


# Inputs 
first_image = tifffile.imread("./S1_DATA/20190603.tif")
second_image = tifffile.imread("./S1_DATA/20190615.tif")

# Detections
change_map = detect_changes(first_image, second_image, filter_size=(2,8), contamination=0.02)


# Plot
cmap = plt.get_cmap('gray')  # 3 levels -1, 0, 1
plt.imshow(change_map, cmap=cmap, vmin=-1, vmax=1)
plt.axis('off')
legend_labels = ['Disappearance (-1)', 'No Change (0)', 'Appearance (1)']
colors = ['black', 'gray', 'white']  # 
legend_elements = [plt.Line2D([0], [0], marker='s', color='w', label=label,
                                 markerfacecolor=color, markersize=10) for label, color in zip(legend_labels, colors)]

plt.legend(handles=legend_elements, loc='upper right', title='Change Categories')

plt.title('Change Detection Map')
plt.show()
```
